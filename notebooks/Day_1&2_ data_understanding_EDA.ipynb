```python
train_clean = train.copy()
```

```python
missing_percent = (train_clean.isnull().sum() / len(train_clean)) * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)
missing_percent.head(10)
```

drop those columns having more than 80% missing values

```python
train_clean = train_clean.drop(columns=["PoolQC", "MiscFeature", "Alley", "Fence"], errors="ignore")
print(train_clean.shape)
```

```python
cat_col = train_clean.select_dtypes(include="object").columns
for col in cat_col:
    train_clean[col] = train_clean[col].fillna("None")

num_col = train_clean.select_dtypes(include=["int64", "float64"]).columns
for col in num_col:
    train_clean[col] = train_clean[col].fillna(train_clean[col].median())
```

```python
train_clean["SalePrice"] = np.log1p(train_clean["SalePrice"])
```

```python
train_clean = pd.get_dummies(train_clean, drop_first=True)
```

get_dummies do the one hot encoding and drop_first avoids multi collinearity

```python
train_clean.to_csv("../data/train_clean.csv", index=False)
```
